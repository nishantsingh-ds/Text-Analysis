```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
import nltk
import re
import os
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent = SentimentIntensityAnalyzer()
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

finaldata=[]
superfinaldata=[]
for i in links:
    
    print("LINK: "+str(i)) 
    
    req = requests.get(i,headers=headers)
    content = req.content # Get the content
    soup = BeautifulSoup(content,'html.parser')
    paragraph = soup.find_all('p')
    title = soup.find_all('title')
    data=title+paragraph
     
    
    index=links.index(i)
    index=index+1
    f = open("%s.txt" % index, "w",encoding='utf-8')

    # traverse paragraphs from soup
    for i in data:
        sum = i.get_text()
        f.writelines(sum)



    f = open("%s.txt" % index, 'r',encoding='utf-8')
    text=f.read()
    f.close()
    
    
    sentences = nltk.sent_tokenize(text)
    wordss = nltk.word_tokenize(text)
    
    from nltk.stem import WordNetLemmatizer
    import re
    lemmatizer=WordNetLemmatizer()

    corpus1=[]
    for i in range(len(sentences)):
        words = nltk.word_tokenize(sentences[i])
        words = re.sub('[^a-zA-Z]', ' ', sentences[i])
        words = words.lower()
        words = words.split()
        words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
        words = ' '.join(words)
        corpus1.append(words)    
    
    #Clean Sentences
    total_sent=0
    for i in range(len(corpus1)):
        i+=1
        total_sent=i   
    print("Total number of Clean Sentences: "+str(i))
    
    #Clean Words
    clean_words=[]
    for i in range(len(corpus1)):
        clean=nltk.word_tokenize(corpus1[i])
        clean_words.append(clean)
    total_clean_words=get_all_elements_in_list_of_lists(clean_words)
    print("Total number of Clean Words: "+str(total_clean_words))
    
    neutral = [(sent.polarity_scores(i)['neu']) for i in wordss]
    negative=[(sent.polarity_scores(i)['neg']) for i in wordss]
    positive=[(sent.polarity_scores(i)['pos']) for i in wordss]
    
    #NEUTRAL SCORE
    neutral_score=0
    for i in range(len(neutral)):
        if neutral[i]==1.0:
            neutral_score=neutral_score+1
    print("Neutral Score: "+ str(neutral_score))
    
    #POSITIVE SCORE
    positive_score=0
    for i in range(len(positive)):
        if positive[i]==1.0:
            positive_score=positive_score+1    
    print("Positive Score: "+ str(positive_score))
    
    #NEGATIVE SCORE
    negative_score=0
    for i in range(len(negative)):
        if negative[i]==1.0:
            negative_score=negative_score+1    
    print("Negative Score: "+ str(negative_score))
    
    #POLARITY SCORE
    numerator= positive_score-negative_score
    denominator= (positive_score+negative_score)+0.000001
    polarity_score = round(numerator/denominator,2)
    print("Polarity Score: "+ str(polarity_score)) 
    
    #SUBJECTIVITY SCORE
    numerator=positive_score +negative_score
    denominator= total_clean_words + 0.000001
    subjectivity_score = round(numerator/denominator,2)
    print("Subjectivity Score: "+ str(subjectivity_score))
    
    #AVERAGE SENTENCE LENGTH OR AVERAGE NUMBER OF WORDS PER SENTENCE
    total_clean_sentences= total_sent
    average_sentence_length= round(total_clean_words/total_clean_sentences,2)
    print("Average Sentence Length: "+ str(average_sentence_length))
    
    #SYLLABLES COUNT
    count=0
    syllables_count=[]
    for i in range(len(clean_words)):    
        for j in range(len(clean_words[i])):
            for w in clean_words[i][j]:
                if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U'):
                    count=count+1
            syllables_count.append(count)
            count=0
    
    #SYLLABLE PER WORD
    total_syllable_count=0
    for i in range(len(syllables_count)):
        total_syllable_count+=syllables_count[i]
    print("Syllables Count: "+str(total_syllable_count))
    
    #COMPLEX WORD COUNT
    complex_word_count=0
    for i in range(len(syllables_count)):
        if syllables_count[i]>2:
            complex_word_count+=1
    print("Complex Word Count: "+str(complex_word_count))   
    
    #PERCENTAGE OF COMPLEX WORD
    percentage_of_complex_words= complex_word_count/total_clean_words
    percentage_of_complex_words=round(percentage_of_complex_words*100,2)
    print("Percentage of Complex Word: "+ str(percentage_of_complex_words))
    
    #FOG INDEX
    fog_index=round(0.4*(average_sentence_length + percentage_of_complex_words),2)
    print("Fog Index: "+ str(fog_index))
    
    sum_of_total_characters=0
    for i in range(len(clean_words)):    
        for j in range(len(clean_words[i])):
            for k in range(len(clean_words[i][j])):
                sum_of_total_characters+=1
    print("Sum of Total Characters: "+ str(sum_of_total_characters))
    
    #AVERAGE WORD LENGTH
    avg_word_length= round(sum_of_total_characters/total_clean_words,2)
    print("Average Word Length: "+ str(avg_word_length))
    
    #PERSONAL PRONOUNS
    pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    pronouns = pronounRegex.findall(text)
    pronouns=len(pronouns)
    print("Personal Pronouns: "+ str(pronouns))
    print(" ")
    
    finaldata=[positive_score,negative_score,polarity_score,subjectivity_score,average_sentence_length,percentage_of_complex_words,fog_index,average_sentence_length,complex_word_count,total_clean_words,total_syllable_count,pronouns,avg_word_length]
    superfinaldata.append(finaldata)
    #print(positive_score,negative_score,polarity_score,subjectivity_score,average_sentence_length,percentage_of_complex_words,fog_index,average_sentence_length,complex_word_count,total_clean_words,total_syllable_count,pronouns,avg_word_length)
    #print(" ")
    #print(superfinaldata)
    print(" ")
```
